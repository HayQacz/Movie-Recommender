# 01 - MovieLens Data Exploration and Preprocessing
## 1. Introduction
This Jupyter Notebook serves as the initial step for our Movie Recommendation System project. Its primary goals are:
* To load the raw MovieLens 1M dataset (users, movies, and ratings).
* To perform an extensive Exploratory Data Analysis (EDA) to understand the characteristics and distributions of the data.
* To merge the different datasets into a unified DataFrame for easier analysis and modeling.
* To perform initial feature engineering, specifically handling movie genres.
* To save the processed data for subsequent modeling steps.

Understanding the data thoroughly is crucial for building effective machine learning models.

## 2. Setup and Imports
We begin by importing all necessary libraries and configuring plotting styles for better visualization.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Configure seaborn for better aesthetics
sns.set_style("whitegrid")
# Set default figure size and font size for plots
plt.rcParams['figure.figsize'] = (12, 7)
plt.rcParams['font.size'] = 12
plt.rcParams['axes.titlesize'] = 16
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
plt.rcParams['legend.fontsize'] = 12
```

## 3. Data Loading
The MovieLens 1M dataset consists of three main files: users.dat, movies.dat, and ratings.dat. These files are delimited by :: and do not contain headers, so we'll specify column names manually. The movies.dat file requires latin-1 encoding due to special characters in movie titles.

```python
# Define the path to the raw data directory
# This assumes the notebook is in 'notebooks/' and raw data in 'data/raw/ml-1m/'
data_path = '../data/raw/ml-1m/'

# Construct full file paths
movies_file = os.path.join(data_path, 'movies.dat')
ratings_file = os.path.join(data_path, 'ratings.dat')
users_file = os.path.join(data_path, 'users.dat')

print(f"File paths:\nMovies: {movies_file}\nRatings: {ratings_file}\nUsers: {users_file}\n")

# Load Users data
users_df = pd.read_csv(users_file, sep='::', engine='python',
                       names=['UserID', 'Gender', 'Age', 'Occupation', 'Zip-code'])
print("--- Users DataFrame Head ---")
print(users_df.head())
print("\n--- Users DataFrame Info ---")
users_df.info()
print("\n" + "="*50 + "\n")

# Load Movies data
movies_df = pd.read_csv(movies_file, sep='::', engine='python', encoding='latin-1',
                        names=['MovieID', 'Title', 'Genres'])
print("--- Movies DataFrame Head ---")
print(movies_df.head())
print("\n--- Movies DataFrame Info ---")
movies_df.info()
print("\n" + "="*50 + "\n")

# Load Ratings data
ratings_df = pd.read_csv(ratings_file, sep='::', engine='python',
                         names=['UserID', 'MovieID', 'Rating', 'Timestamp'])
print("--- Ratings DataFrame Head ---")
print(ratings_df.head())
print("\n--- Ratings DataFrame Info ---")
ratings_df.info()
print("\n" + "="*50 + "\n")
```

## 4. Exploratory Data Analysis (EDA)
In this section, we'll perform an initial exploration of the loaded datasets to understand their basic statistics, distributions, and identify any potential issues like missing values.

### 4.1. Missing Values Check

```python
print("--- Missing values in Users DataFrame ---")
print(users_df.isnull().sum())
print("\n--- Missing values in Movies DataFrame ---")
print(movies_df.isnull().sum())
print("\n--- Missing values in Ratings DataFrame ---")
print(ratings_df.isnull().sum())
```

### 4.2. Unique Values and Distributions
Let's look at the unique values and their counts for categorical columns, and the distribution of ratings.

```python
print("\n--- Unique values and counts for Users DataFrame categorical columns ---")
for col in ['Gender', 'Age', 'Occupation']:
    print(f"\nColumn: {col}")
    print(f"Number of unique values: {users_df[col].nunique()}")
    print(users_df[col].value_counts().sort_index())

print("\n--- Ratings Distribution ---")
print(f"Unique ratings: {ratings_df['Rating'].unique()}")
print("Rating value counts:\n", ratings_df['Rating'].value_counts().sort_index())

print(f"\nNumber of unique users: {users_df['UserID'].nunique()}")
print(f"Number of unique movies: {movies_df['MovieID'].nunique()}")
print(f"Total number of ratings: {len(ratings_df)}")
```

### 4.3. Visualizing Rating Distribution

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Rating', data=ratings_df, palette='viridis')
plt.title('Distribution of Movie Ratings')
plt.xlabel('Rating')
plt.ylabel('Number of Ratings')
plt.show()
```

### 5. Data Merging
To facilitate comprehensive analysis and model building, we'll merge the three DataFrames (ratings_df, movies_df, users_df) into a single, unified DataFrame.

```python
# Merge ratings with movies
df = pd.merge(ratings_df, movies_df, on='MovieID', how='inner')
# Merge the result with users
df = pd.merge(df, users_df, on='UserID', how='inner')

print("--- Merged DataFrame Head ---")
print(df.head())
print("\n--- Merged DataFrame Info ---")
df.info()
print(f"\nShape of the merged DataFrame: {df.shape}")
```

## 6. Further Exploration of Merged Data
Now that we have a consolidated DataFrame, we can perform more insightful analyses.

### 6.1. Top Movies by Number of Ratings
Identifying movies that have received the most ratings.

```python
top_rated_movies_count = df['Title'].value_counts().head(10)
print("\n--- Top 10 Movies by Number of Ratings ---")
print(top_rated_movies_count)

plt.figure(figsize=(12, 7))
sns.barplot(x=top_rated_movies_count.values, y=top_rated_movies_count.index, palette='magma')
plt.title('Top 10 Movies by Number of Ratings')
plt.xlabel('Number of Ratings')
plt.ylabel('Movie Title')
plt.show()
```

### 6.2. Top Movies by Average Rating (with a minimum rating threshold)
To get a more reliable average rating, we'll only consider movies that have received a minimum number of ratings (e.g., 50 ratings) to avoid bias from movies with very few, but high, ratings.

```python
min_ratings_threshold = 50
movie_stats = df.groupby('Title').agg(
    mean_rating=('Rating', 'mean'),
    num_ratings=('Rating', 'count')
).reset_index()

# Filter for popular movies (meeting the minimum ratings threshold)
popular_movies = movie_stats[movie_stats['num_ratings'] >= min_ratings_threshold]
top_average_movies = popular_movies.sort_values(by='mean_rating', ascending=False).head(10)

print(f"\n--- Top 10 Movies by Average Rating (min. {min_ratings_threshold} ratings) ---")
print(top_average_movies)

plt.figure(figsize=(12, 7))
sns.barplot(x='mean_rating', y='Title', data=top_average_movies, palette='cubehelix')
plt.title(f'Top 10 Movies by Average Rating (min. {min_ratings_threshold} ratings)')
plt.xlabel('Average Rating')
plt.ylabel('Movie Title')
plt.xlim(3.5, 5.0)
plt.show()
```

### 6.3. Rating Distribution by Gender


```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Rating', y='Gender', data=df, palette='pastel')
plt.title('Rating Distribution by Gender')
plt.xlabel('Rating')
plt.ylabel('Gender')
plt.show()
```

## 7. Feature Engineering: Handling Genres
The 'Genres' column contains multiple genres separated by |. To make this information usable for machine learning models, we'll convert it into a one-hot encoded format.

```python
# Create a copy to avoid modifying the original movies_df
movies_df_copy = movies_df.copy()

# Extract all unique genres
all_genres = movies_df_copy['Genres'].str.split('|').explode().unique()
print(f"\nTotal unique genres: {len(all_genres)}")
print(all_genres)

# Create one-hot encoded DataFrame for genres
genre_dummies = movies_df_copy['Genres'].str.get_dummies('|')
movies_df_encoded = pd.concat([movies_df_copy, genre_dummies], axis=1)

print("\n--- Movies DataFrame with One-Hot Encoded Genres (Head) ---")
print(movies_df_encoded.head())

# Visualize the popularity of genres
genre_counts = genre_dummies.sum().sort_values(ascending=False)
print("\n--- Number of Movies per Genre (Top 10) ---")
print(genre_counts.head(10))

plt.figure(figsize=(14, 8))
sns.barplot(x=genre_counts.values, y=genre_counts.index, palette='viridis')
plt.title('Number of Movies per Genre')
plt.xlabel('Number of Movies')
plt.ylabel('Genre')
plt.show()
```

## 8. Data Type Conversion
Convert the 'Timestamp' column in the merged DataFrame to a datetime object for potential time-based analysis.

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'], unit='s')
print("\n--- Merged DataFrame after Timestamp conversion ---")
print(df.head())
print(df.info())
```

## 9. Saving Processed Data
Finally, we'll save the merged and genre-encoded DataFrames to the data/processed/ directory. This step is crucial as it allows us to load clean, preprocessed data directly in subsequent notebooks, saving time and computational resources.

```python
# Save the main merged DataFrame
output_merged_path = '../data/processed/merged_movie_data.csv'
df.to_csv(output_merged_path, index=False)
print(f"\nProcessed merged data saved to: {output_merged_path}")

# Save the movies DataFrame with one-hot encoded genres
output_movies_encoded_path = '../data/processed/movies_encoded.csv'
movies_df_encoded.to_csv(output_movies_encoded_path, index=False)
print(f"Movies with encoded genres saved to: {output_movies_encoded_path}")
```

## 10. Conclusion
This notebook successfully loaded, explored, and preprocessed the MovieLens 1M dataset. We gained insights into user demographics, movie characteristics, and rating patterns. The merged and genre-encoded DataFrames are now ready for the next phase: model training and evaluation.
